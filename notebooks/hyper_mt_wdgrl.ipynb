```python
import os
import pandas as pd
from hyper_ploter import join_data, disc_plot, src_trg_plot, wasserstein_plot,\
wassestein_valid_plot, losses_plot, src_trg_valid_plot, losses_valid_plot
# jupyter nbconvert --to pdf --TemplateExporter.exclude_input=True --TemplateExporter.exclude_output_prompt=True hyper_mt_wdgrl.ipynb
```

Les courbes ci dessous représentent les valeurs des losses, accuracies (à la fin de chaque éxécution) en fonction des valeurs des hypers paramètres. Nous introduisons les notations suivalntes:

* $\mathcal{L}_C(S)$ et $\mathcal{L}_C(T)$ les losses de classification sur les domaines soruces et target respectivement
* $\mathcal{A}_C(S)$ et $\mathcal{A}_C(T)$ les précisions de classification sur les domaines soruces et target respectivement
* $\hat{\mathcal{L}}_{wd}$ l'estimation de la distance de Wasserstein obtenue après résolution du problème $\underset{\theta_w}{\max}\{\mathcal{L}_{wd}−γ\mathcal{L}_{grad}\}$

La loss totale devient donc $\mathcal{L} = \mathcal{L}_C(S) + \lambda \hat{\mathcal{L}}_{wd}$

## Variations de $\lambda$

Nous faisons varier $\lambda$ sur l'ensemble des valeurs suivant $\{0.0, 0.001, 0.01, 0.1, 1.0, 2.0, 5.0, 10.0\}$. Pour une meilleure lisibilité, nous avons uyilisé le logarithme en base 10 de $\lambda$ pour les tracés.

```python
logdir = '../log/'
hyper_name = 'lambda'
xlabel = '$\log(\\' + hyper_name +')$'
data_folder = os.path.join(logdir, f'micro_tcga_wdgrl_{hyper_name}_2020-10-01_18-50-59')

# read meta data
meta_df = pd.read_csv(os.path.join(data_folder, 'meta'), 
                      index_col=0, header=0, names=['param', 'folder'])
n_params = meta_df.shape[0]
```

### Discriminator

```python
df = join_data(data_folder, 'disc', meta_df)
disc_plot(df, xlabel=xlabel)
```

## WDGRL Micorarray -> TCGA

### Training

```python
df = join_data(data_folder, 'train', meta_df)
src_trg_plot(df, xlabel=xlabel)
wasserstein_plot(df, xlabel=xlabel)
losses_plot(df, xlabel=xlabel)
```

### Validation

```python
df = join_data(data_folder, 'valid', meta_df)
src_trg_valid_plot(df, xlabel=xlabel)
wassestein_valid_plot(df, xlabel=xlabel)
losses_valid_plot(df, xlabel=xlabel)
```
